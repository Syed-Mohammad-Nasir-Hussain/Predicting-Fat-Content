```python
import statsmodels.formula.api as smf # Importing Required Libraries
```

```python
m1=smf.ols('FAT~Waist',data=x).fit() # Building Simple Linear Regression model on X data set
```

```python
m1.params # Gives Parameters info 
```

```python
m1.summary() # Checking accuracy and significance
```

```python
pred=m1.predict(x) # Predicting the output
```
